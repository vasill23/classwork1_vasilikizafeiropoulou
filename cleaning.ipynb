```python
from google.colab import drive
drive.mount('/content/gdrive')

import pandas as pd
import os
from glob import glob

articles = '/content/gdrive/MyDrive/διαφθορα'  # <-- άλλαξε αυτό

csv_files = glob(os.path.join(articles, '*.csv'))

print("Βρέθηκαν αρχεία:", csv_files)

```

```python
df_main = pd.DataFrame()

for file in csv_files:
    try:
        df_temp = pd.read_csv(file,delimiter=';', on_bad_lines='skip')  # Αν έχει προβληματικές γραμμές, θα τις αγνοήσει
        df_temp['/content/gdrive/MyDrive/διαφθορα'] = os.path.basename(file)   # Προσθήκη στήλης με όνομα αρχείου (προαιρετικό αλλά χρήσιμο)
        df_main = pd.concat([df_main, df_temp], ignore_index=True)
    except Exception as e:
        print(f"Πρόβλημα στο αρχείο {file}: {e}")

```

```python
df_main
```

```python
df_main= df_main[df_main["type"] == "web"].drop_duplicates(subset="title")
```

```python
df_main
```

```python
df_main = df_main.dropna(subset=["date"]).reset_index(drop=True)
df_main = df_main.dropna(subset=["text"]).reset_index(drop=True)
df_main = df_main.dropna(subset=["title"]).reset_index(drop=True)
```

```python
df_main
```

```python
df_main['text'] = df_main['text'].str.replace('\n', ' ').replace('\r', ' ').replace('\t', ' ')
```

```python
df_main.groupby("source")['title'].count().reset_index().sort_values("title", ascending=False).iloc[0:20, :]
```

```python
# Βήμα 1: Βρες τους 20 top sources
top_sources = (
    df_main.groupby("source")['title']
    .count()
    .reset_index()
    .sort_values("title", ascending=False)
    .iloc[0:20, :]['source']
)

# Βήμα 2: Φίλτραρε το αρχικό DataFrame για να κρατήσεις μόνο τα άρθρα από αυτούς
df_top = df_main[df_main['source'].isin(top_sources)]
```
